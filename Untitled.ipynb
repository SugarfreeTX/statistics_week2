```python
file = open('data.txt')
word_counts = {}
for row in file: 
    for word in row.split():
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1
file.close()
```

```python

for word in word_counts:
    print(word, word_counts[word])
```

```python
for word in sorted(word_counts, key=lambda w: word_counts[w], reverse=True):
    print(word, word_counts[word])
```
