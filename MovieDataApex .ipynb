```python
import datetime as dt
def make_date(date_str):
    m, d, y = date_str.split('/')
    m = int(m)
    d = int(d)
    y = int(y)
    if y > 13:
        y += 1900
    else:
        y += 2000
    return dt.datetime(y, m, d)
```

```python
print(file["US Gross"].unique())
print(file["Worldwide Gross"].unique())
```

```python
movie = pd.read_csv("moviedata.csv", sep='\t', na_values=['Unknown', 'Unkno'], parse_dates=[0], date_parser=make_date)
```
