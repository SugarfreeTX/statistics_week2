```python
def convert_to_int(text):
    try:
        return int(text)
    except:
        return None
    
```

```python
file = open("moviedata.csv")
data = []
print(file.readline())
for row in file:
    row = row.split('\t')
    for i in range(3, 6):
        row[i] = convert_to_int(row[i])
    data.append(row)
```

```python
# US Gross average. 
total = 0.0
count = 0.0
for row in data:
    try:
        total += row[4]
        count += 1
    except:
        pass
print(total / count)
```

```python
# Budget over 20 mil
# Just doing a count. 
big_budget = 0.0
for row in data:
    if row[3] > 20000000:
        big_budget += 1
print(big_budget)
```

```python
# How many movies were released by each film distrubutor. Use dictionary.
distributors = {}
for row in data:
    distributor = row[2]
    if distributor in distributors:
        distributors[distributor] += 1
    else:
        distributors[distributor] = 1

for distributor in distributors:
    print(distributor, distributors[distributor])
```

```python
# Convert all to a float. 
def convert_to_float(record):
    try:
        return float(record) # convert to float value. 
    except:
        return record # return a string
```

```python
quake = open("quakedata.csv")
print(quake.readline().split(','))
quake_data = []
for row in quake: 
    row = row.split(',')
    row = convert_to_float(row)
    quake_data.append(row)
```

```python
print(len(quake_data))
```

```python
# What is the average magnitude. 
# put this in a column. 
all_mag = [row[4] for row in quake_data]
```

```python
all_mags = []
for row in all_mag:
    row = convert_to_float(row)
    all_mags.append(row)

```

```python
total = sum(all_mags)
mean = total * 1.0 / len(all_mags)
print(mean)
```

```python
def make_datetime(date_str):
    year = int(date_str[:4])
    month = int(date_str[5:7])
    day = int(date_str[8:10])
    hour = int(date_str[11:13])
    minute = int(date_str[14:16])
    second = int(date_str[17:19])
    return dt.datetime(year, month, day, hour, minute, second)
```

```python
text = open("data.txt")
word_countx = {}
for row in text:
    for word in row.split():
        word = word.lower()
        word = word.strip(',.!?- ')
        if word in STOP_WORDS:
            continue # means to go to the next word.
        if word in word_countx:
            word_countx[word] += 1
        else:
            word_countx[word] = 1
text.close()
            
        
```

```python
# sort the words and store them in a list. 
sorted_words = sorted(word_countx, key= lambda w: word_countx[w], reverse =True)
```

```python
for words in sorted_words:
    count = word_countx[words]
    if count > 1:
        print(words, count)
```
